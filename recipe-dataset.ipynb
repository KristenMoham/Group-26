```python
#numberSteps = df[(df['n_steps'] == 6)]
mainIngredient = df[['name','ingredients','n_ingredients','n_steps','minutes']]
#print(mainIngredient.head())
mIng = mainIngredient.values.tolist()
sIng = []
#print(smIng)
#print(mIng[2][1][2:])
```

```python
#Quick Sort
```

```python
#Shell Sort
n = len(mIng)
gap = n//2
while gap > 0:
        for i in range(gap, n):
            temp = mIng[i]
            j = i            
            temp_ingredient = temp[1][2:10]
            while j >= gap:
                prev = mIng[j - gap]
                prev_ingredient = prev[1][2:10]
                if prev_ingredient > temp_ingredient:
                    mIng[j] = prev
                j -= gap
            mIng[j] = temp
            
        if(gap == 2):
            gap = 1
        else:
            gap //= 2
        print(gap)

print(mIng[1][1])
print(mIng[2][1])
print(mIng[3][1])
print(mIng[4][1])
print(mIng[5][1])
print(mIng[6][1])
print(mIng[7][1])
print(mIng[8][1])
print(mIng[9][1])
```
